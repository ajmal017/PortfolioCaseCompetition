```python
from IPython.display import display
import pandas as pd
from datetime import datetime
import yfinance as yf
from matplotlib import pyplot as plt
```

# Defining the two Portfolios

```python
optimal_portfolio = pd.read_pickle('Optimal Portfolio.pkl')['Optimal Shares']
current_portfolio = pd.read_pickle('Current Portfolio.pkl')['Shares Holding']
```

## Getting Historical Close Prices

```python
end = datetime.now().strftime("%Y-%m-%d")
start = '2019-11-10'


historical_data_optimal = yf.download(list(optimal_portfolio.index),
                             start,
                             end)['Close']

historical_data_current = yf.download(list(current_portfolio.index),
                             start,
                             end)['Close']

```

```python
pnl_optimal = historical_data_optimal - historical_data_optimal.iloc[0, :]
pnl_optimal = pnl_optimal * optimal_portfolio

pnl_current = historical_data_current - historical_data_current.iloc[0, :]
pnl_current = pnl_current * current_portfolio
```

## Plotting Absolute Returns

```python
plt.figure(figsize=(15, 5))
(100*pnl_optimal.sum(axis=1)/(historical_data_optimal.iloc[0, :]*optimal_portfolio).sum()).plot(figsize=(15, 5), title='Optimal PnL'); plt.grid(); plt.ylabel('%')

plt.figure(figsize=(15, 5))
(100*pnl_current.sum(axis=1)/(historical_data_current.iloc[0, :]*current_portfolio).sum()).plot(figsize=(15, 5), title='Current PnL'); plt.grid(); plt.ylabel('%')

plt.show()
```

## Evaluating Sharpe Ratios

```python
Opt = pnl_optimal.sum(axis=1)/(historical_data_optimal.iloc[0, :]*optimal_portfolio).sum()
Cur = pnl_current.sum(axis=1)/(historical_data_current.iloc[0, :]*current_portfolio).sum()
```

```python
print('Optimal Portfolio Sharpe: ', Opt.mean()/Opt.std())
print('Current Portfolio Sharpe: ', Cur.mean()/Cur.std())
```
